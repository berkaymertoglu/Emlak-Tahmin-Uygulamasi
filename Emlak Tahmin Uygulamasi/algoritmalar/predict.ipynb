```python
import pandas as pd
import numpy as np
import joblib
from sklearn.preprocessing import LabelEncoder
```

```python
dt_model = joblib.load("decision_tree_model.pkl")
rf_model = joblib.load("random_forest_model.pkl")
lr_model = joblib.load("linear_regression_model.pkl")
svr_model = joblib.load("support_vector_regression.pkl")
knn_model = joblib.load("k_nearest_neighbors_model.pkl")
gbm_model = joblib.load("gradient_boosting_machine_model.pkl")
ridge_model = joblib.load("ridge_regression_model.pkl")
lasso_model = joblib.load("lasso_regression_model.pkl")
sequential_model = joblib.load("sequential_model.pkl")
```

```python
def balkon_durum(balkon_durumu):
    if balkon_durumu.lower() == "var":
        return 1
    elif balkon_durumu.lower() == "yok":
        return 0
    else:
        print("Hata: Geçersiz Giriş!")
```

```python
def bulunulan_kat(bulundugu_kat):
    index_no = int(kat[kat["Bulunduğu Kat"] == bulundugu_kat].index.values)
    return index_no
```

```python
def bina_yas(binanin_yasi):
    index_no = int(yas[yas["Binanın Yaşı"] == binanin_yasi].index.values)
    return index_no
```

```python
def isitma_tip(isitma_tipi):
    index_no = int(isitma[isitma["Isıtma Tipi"] == isitma_tipi].index.values)
    return index_no
```

```python
def oda_sayi(oda_sayisi):
    index_no = int(oda[oda["Oda Sayısı"] == oda_sayisi].index.values)
    return index_no
```

```python
def esya_durum(esya_durumu):
    if esya_durumu.lower() == "eşyalı":
        return 1
    elif esya_durumu.lower() == "boş":
        return 0
    else:
        print("Hata: Geçersiz Giriş!")
```

```python
def site_ici(site_icerisinde):
    if site_icerisinde.lower() == "evet":
        return 1
    elif site_icerisinde.lower() == "hayır":
        return 0
    else:
        print("Hata: Geçersiz Giriş!")  
```

```python
def tahmin_deger(balkon_durumu, kat_sayisi, binanin_yasi, brüt_m2, bulundugu_kat, esya_durumu, isitma_tipi, net_m2, oda_sayisi, site_icerisinde):
    pred = pd.DataFrame(data =
                        {"Balkon Durumu": [balkon_durumu], "Kat Sayısı": [kat_sayisi],
                         "Binanın Yaşı": [binanin_yasi], "Brüt Metrekare": [brüt_m2],
                         "Bulunduğu Kat": [bulundugu_kat], "Eşya Durumu": [esya_durumu],
                         "Isıtma Tipi": [isitma_tipi], "Net Metrekare": [net_m2],
                         "Oda Sayısı" : [oda_sayisi], "Site İçerisinde": [site_icerisinde]})
    return pred
```

```python
def tahmin_et(pred):
    print("Decision Tree Result: " + str(int(dt_model.predict(pred))).strip("[]") + " TL")
    print("Random Forest Result: " + str(int(rf_model.predict(pred))).strip("[]") + " TL")
    print("Linear Regression Result: " + str(int(lr_model.predict(pred))).strip("[]") + " TL")
    print("Support Vector Regression Result: " + str(int(svr_model.predict(pred))).strip("[]") + " TL")
    print("K-Nearest Neighbors Result: " + str(int(knn_model.predict(pred))).strip("[]") + " TL")
    print("Gradient Boosting Machine Result: " + str(int(gbm_model.predict(pred))).strip("[]") + " TL")
    print("Ridge Regression Result: " + str(int(ridge_model.predict(pred))).strip("[]") + " TL")
    print("Lasso Regression Result: " + str(int(lasso_model.predict(pred))).strip("[]") + " TL")
    print("Sequential Model Result: " + str(int(sequential_model.predict(pred))).strip("[]") + " TL")
```

```python
tahmin_edilen_ev = tahmin_deger(balkon_durum("yok"), 3, bina_yas("21 Ve Üzeri"), 120, bulunulan_kat("1.Kat"), esya_durum("Boş"), isitma_tip("Sobalı"), 90, oda_sayi("Stüdyo"), site_ici("hayır"))
tahmin_et(tahmin_edilen_ev)
```
