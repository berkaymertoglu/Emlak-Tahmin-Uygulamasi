```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url_list = []
prices_list = []
propTitles = []
propValues = []
```

```python
for i in range (1,25):
    url = "https://www.emlakjet.com/kiralik-daire/istanbul/" + str(i)
    r = requests.get(url)
    source = BeautifulSoup(r.content, "lxml")
    
    urls = source.find_all("div", attrs = {"class":"_3qUI9q"})
    for url in urls:
        url_home = "https://www.emlakjet.com/kiralik-daire/istanbul/" + url.a.get("href")
        url_list.append(url_home)
        print(url_home)
        
        r_home = requests.get(url_home)
        source_home = BeautifulSoup(r_home.content,"lxml")
        
        properties = source_home.find_all("div", attrs = {"class":"_35T4WV"})
        for prop in properties:
            prop_title = prop.find("div", attrs = {"class": "_1bVOdb"}).text
            prop_value = prop.find_all("div", attrs = {"class": "_1bVOdb"})[1].text
            propTitles.append(prop_title)
            propValues.append(prop_value)
            
    prices = source.find_all("div", attrs = {"class": "_3Q-7xT"}) 
    
    for price in prices:
        prices_list.append(price.text)
        print(price.text)
        
        
print(str(len(url_list)) + " Adet link bulundu")
print(str(len(prices_list)) + " Adet fiyat bulundu")
print(str(len(propTitles)) + " Adet ozellik bulundu")
print(str(len(propValues)) + " Adet ozellik verisi bulundu")
```

```python
df_urls = pd.DataFrame()
df_urls["urls"] = url_list
df_urls["prices"] = prices_list
```
